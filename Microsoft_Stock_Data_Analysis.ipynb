**Microsoft Stock Data Analysis**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Microsoft Dataset.csv')
```

**Data Cleaning**

```python
df.isnull().sum()
```

```python
Date_time = pd.to_datetime(df['Date'])
df['Date'] = Date_time
```

```python
df.info()
```

**Stock Price Visualisation**

```python
#Line Plot
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(df['Date'], df['Close'] , color= 'blue')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Microsoft Stock Price')
plt.xlabel('Date')
plt.ylabel('Stock Price in USD($)')
plt.grid()
plt.show()

```

```python
#Bar Plot
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(df['Date'], df['Close'] , color= 'blue')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Microsoft Stock Price')
plt.xlabel('Date')
plt.ylabel('Stock Price in USD($)')
plt.grid()
plt.show()
```

**Market Capitisation Visual**

```python
df['MarketCap'] = df['Open'] * df['Volume']
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax.bar(df['Date'], df['MarketCap'] , color= 'Red')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Microsoft Stock MarketCap')
plt.xlabel('Date')
plt.ylabel('MarketCap')
plt.grid()
plt.show()
```

**Stock Price Volatility**

```python
df['Volatility'] = (df['Close']/ df['Close'].shift(1))-1
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax.bar(df['Date'], df['Volatility'] , color= 'Brown')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Microsoft Stock Price Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax.plot(df['Date'], df['Volatility'] , color= 'Brown')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Microsoft Stock Price Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid()
plt.show()
```

```python

df['Volatility'].hist(bins=1000, color='skyblue', edgecolor='black')
plt.xlabel('Volatility')
plt.ylabel('Frequency')
plt.title('Volatility Frequency Distribution')
plt.show()
```

```python
df['Cumulative Return'] = (1 + df['Volatility']).cumprod()
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax.plot(df['Date'], df['Cumulative Return'] )
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Microsoft Stock Price Cumulative return')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.grid()
plt.show()
```

**Moving Averages **

```python
df["MA for 10 days"] = df['Open'].rolling(10).mean()
df["MA for 20 days"] = df['Open'].rolling(20).mean()
df["MA for 50 days"] = df['Open'].rolling(50).mean()
df["MA for 100 days"] = df['Open'].rolling(100).mean()

df[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days', 'MA for 100 days']].plot(subplots=False, figsize=(12, 5))
plt.title('Microsoft Stock: Adjusted Close Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
